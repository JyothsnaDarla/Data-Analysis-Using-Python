```python
#list compresion
#newlist=[output for]
l=[4,5,33,6,5.3]
d=[i*i for i in l]
print(d)
```

```python
#list comprehension
#newlist=[output for condition]
l=[4,5,33,6,5.3]
d=[i*i for i in l if(i%2==0)]
print(d)
```

```python
#list comprehension
#newlist=[output for condition]
l=[4,5,33,65,5.3]
d=[i*i for i in l if(i>50)]
print(d)
```

```python
#list comprehension
#newlist=[(st of if) if (con) else (st of else)]
l1=[23,67,12,34,69]
res=[i*i if i>50 else 'X' for i in l1]
print(res)
```
