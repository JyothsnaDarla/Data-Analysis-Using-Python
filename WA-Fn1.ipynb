```python
feature_cols=['Age','EmployeeNumber','TotalWorkingYears','YearsWithCurrManager','Education']
features=df[feature_cols]
features
```

```python
Target=df.Attrition
```

```python
from sklearn.ensemble import RandomForestClassifier
ml=RandomForestClassifier(n_estimators=100)
```

```python
ml.fit(features,Target)
```

```python
c=input("enter age ")
d=input("enter EmployeeNumber ")
e=input("enter TotalWorkingYears ")
b=input("enter 	YearsWithCurrManager ")
f=input("enter Education ")
pred=ml.predict([[c,d,e,b,f]])
if pred =='Yes':
  print("not working")
else:
  print("hiried")
```

```python
import os
output_dir="tree_visualizations"
os.makedirs(output_dir,exist_ok=True)
```

```python
ml.estimators_
```

```python
for i,tree in enumerate(ml.estimators_):
  tree_dot_file=os.path.join(output_dir,f"tree_{i}.dot")
  tree_png_file=os.path.join(output_dir,f"tree_{i}.png")
```

```python
from sklearn.tree import export_graphviz
export_graphviz(tree, out_file=tree_dot_file, feature_names=['Age','EmployeeNumber','TotalWorkingYears','YearsWithCurrManager','Education'],
                    class_names=[str(cls) for cls in ml.classes_], filled=True, rounded=True)
```

```python
command= f"dot -Tpng {tree_dot_file} -o {tree_png_file}"
os.system(command)
print(f"Tree{i}visualization saved to {tree_png_file}")
```
