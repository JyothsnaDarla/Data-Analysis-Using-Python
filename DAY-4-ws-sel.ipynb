```python
import selenium
from selenium import webdriver
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
from webdriver_manager.chrome import ChromeDriverManager
import pandas as pd
```

```python
#define options and set browser capabilities
options=webdriver.ChromeOptions()
#options.add_argument('--seme-option')
```

```python
#create webDriver instance with options
driver=webdriver.Chrome(options=options)
```

```python
#navigate to a website
driver.get('https://www.myntra.com/')
```

```python
search=driver.find_element(By.XPATH,".//input [@class='desktop-searchBar']")
```

```python
search.send_keys("Cardigan Tops")
```

```python
search.send_keys(Keys.ENTER)
```

```python
brd=driver.find_elements(By.XPATH,".//h3[@class='product-brand']")
brds=[i.text for i in brd]
print(brds)
print(len(brds))
```

```python
des=driver.find_elements(By.XPATH,".//h4[@class='product-product']")
desc=[i.text for i in des]
print(desc)
print(len(desc))
```

```python
price=driver.find_elements(By.XPATH,".//span[@class='product-discountedPrice']")
prices=[i.text for i in price]
print(prices)
print(len(prices))
```

```python
df=pd.DataFrame(columns=["brd","des","price"])
df
```

```python
df['brd']=brds
df['des']=desc
df['price']=prices
```

```python
df
```

```python
df.to_csv("df.csv")
```
